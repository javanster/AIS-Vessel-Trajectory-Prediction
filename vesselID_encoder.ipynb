```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

train_df = pd.read_csv('data/original_data/ais_train.csv', sep='|')
test_df = pd.read_csv('data/original_data/ais_test.csv')
```

```python
label_encoder = LabelEncoder()

label_encoder.fit(train_df['vesselId'])

train_df['vesselId'] = label_encoder.transform(train_df['vesselId'])
test_df['vesselId'] = label_encoder.transform(test_df['vesselId'])

train_df.head()
```

```python
train_df.to_csv('data/processed_data/train_encoded_vessel_ID.csv', index=False)
test_df.to_csv('data/processed_data/test_encoded_vessel_ID.csv', index=False)
```
